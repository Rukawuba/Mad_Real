```python
import pandas as pd

# Assuming 'ds' is your dataset and 'price' and 'area' are column names
# Group data by area and calculate mean price
mean_prices_by_area = ds.groupby('address')['price'].mean()

# Find the area with the highest mean price
highest_mean_price_area = mean_prices_by_area.idxmax()

# Get the highest mean price itself
highest_mean_price = mean_prices_by_area.max()

print(f"Area with the highest mean price: {highest_mean_price_area}")
print(f"Highest mean price: {highest_mean_price:.2f}")  # Format price with 2 decimals
```
